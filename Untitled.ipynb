```python
import numpy as np
import pandas as pd
```

```python
rev = pd.read_csv("data/SKU Sold Detail.csv")
rev.head()
```

```python
rev['Revenue'] = rev['Revenue'].replace(" ",np.nan)
rev = rev[rev['Revenue'].notnull()]
rev = rev.reset_index()[rev.columns]
rev['Revenue'] = rev['Revenue'].astype(float)
```

```python
# Grouping columns by Region, Location & Year
groupbyYear = rev.groupby([(rev.Quarter),(rev.Country)]).agg(
    Revenue = ('Revenue','sum'),).reset_index()
groupbyYear
```

```python
import matplotlib.pyplot as plt
# set width of bars
w = 0.5

# Figure Size
fig = plt.figure(figsize =(10, 8))

# Define x-axis
year = groupbyYear['Quarter']

# Define y-axis
Revenue = groupbyYear['Revenue']



# Make the plot cases vs death
plt.bar(year, Revenue, 0.4, color='#89CFF0', label='New Cases')

# Add the labels
plt.xlabel('Quarter')
plt.ylabel('Revenue')
plt.title("Revenue by Quarter", fontweight='bold')
 
# Create legend & Show graphic
plt.legend(loc='upper left')
plt.show()
```

```python
groupbyYear = rev.nlargest(n=50, columns=['Revenue'])
groupbyYear.head(5)
```
