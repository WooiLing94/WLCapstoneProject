```python
#for data preprocessing
import pandas as pd

#for matrix math
import numpy as np

#for plotting
import seaborn as sns                #style setting
import matplotlib.pyplot as plt      #graph plotting

#Standardise data value
from sklearn.preprocessing import StandardScaler
```

```python
#Load Dataset
sold = pd.read_csv('data/SKU Sold Detail.csv')
sold
```

```python
#detect missing value in data frame
sold.isnull().any()
```

```python
numericalData = sold[["Country","Revenue_Apr21","Revenue_Oct21","Revenue_Jan22"]]
numericalData.head()
```

```python
#format the figure size to 20rows,10cols
fig, ax = plt.subplots(figsize = (20,10))

#plot graph by Revenue Apr21
plt.subplot(221)
sns.boxplot(x = numericalData["Revenue_Apr21"])
plt.title("Revenue Apr21", fontsize = 20)

#plot graph by Revenue Oct21
plt.subplot(222)
sns.boxplot(x = numericalData["Revenue_Oct21"])
plt.title("Revenue Oct21", fontsize = 20)

#plot graph by Revenue Jan22
plt.subplot(223)
sns.boxplot(x = numericalData["Revenue_Jan22"])
plt.title("Revenue Jan22", fontsize = 20)


plt.tight_layout()
plt.show()
```

```python
sorted(numericalData['Revenue_Apr21'])
q1, q3, m= np.percentile(numericalData['Revenue_Apr21'],[25,75,50])
iqr = q3 - q1
lowerb = q1 - (1.5 * iqr)
upperb = q3 + (1.5 * iqr)
print("Apr21 LowerB: ",lowerb)
print("Apr21 Upper B: ",upperb)

median = numericalData.loc[numericalData['Revenue_Apr21']<= upperb,'Revenue_Apr21'].median()
numericalData.loc[numericalData['Revenue_Apr21']> upperb,'Revenue_Apr21'] = np.nan
numericalData.fillna(median,inplace=True)

sorted(numericalData['Revenue_Oct21'])
q1, q3, m= np.percentile(numericalData['Revenue_Oct21'],[25,75,50])
iqr = q3 - q1
lowerb = q1 - (1.5 * iqr)
upperb = q3 + (1.5 * iqr)
print("Oct21 LowerB: ",lowerb)
print("Oct21 Upper B: ",upperb)

median = numericalData.loc[numericalData['Revenue_Oct21']<= upperb,'Revenue_Oct21'].median()
numericalData.loc[numericalData['Revenue_Oct21']> upperb,'Revenue_Oct21'] = np.nan
numericalData.fillna(median,inplace=True)

sorted(numericalData['Revenue_Jan22'])
q1, q3, m= np.percentile(numericalData['Revenue_Jan22'],[25,75,50])
iqr = q3 - q1
lowerb = q1 - (1.5 * iqr)
upperb = q3 + (1.5 * iqr)
print("Jan22 LowerB: ",lowerb)
print("Jan22 Upper B: ",upperb)

median = numericalData.loc[numericalData['Revenue_Jan22']<= upperb,'Revenue_Jan22'].median()
numericalData.loc[numericalData['Revenue_Jan22']> upperb,'Revenue_Jan22'] = np.nan
numericalData.fillna(median,inplace=True)
```
